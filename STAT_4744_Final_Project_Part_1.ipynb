```python
# Import necessary libraries.
import torch
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt

# Load MNIST data.
data = torchvision.datasets.MNIST(root="./data", train=True, download=True)

# Display 5 MNIST images.
for i in range(5):
    image, label = data[i]

    # Original image
    plt.subplot(2, 5, i+1)
    plt.imshow(image, cmap='gray')
    plt.title(f"{label}")
    plt.axis('off')

# Display inverted MNIST image for comparison.
image, label = data[0]
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.show()

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: 1 - x)
])

inverted = transform(image)

plt.imshow(inverted.squeeze().numpy(), cmap='gray')
plt.axis('off')
plt.show()
```
